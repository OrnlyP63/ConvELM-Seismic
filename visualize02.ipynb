```python
from obspy.clients.syngine import Client
client = Client()
```

```python
st = client.get_waveforms(model="ak135f_5s", network="IU", station="ANMO",
                          eventid="GCMT:C201002270634A")
```

```python
st.plot()
```

```python
st[0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(st[0].data)
```

```python
plt.figure(figsize=(18, 5))
s = st[0]
sp = s[2800:5000]
ss = s[5000:8100]
sb = s[8100:13400]
plt.plot(range(len(s)), s,  c='k')
plt.plot(range(2800, 5000),sp, c='r', label='P-wave')
plt.plot(range(5000, 8100),ss, c='g', label='S-wave')
plt.plot(range(8100, 13400),sb, c='b', label='Surface-wave')
plt.grid()
plt.legend()
plt.savefig('wave.png')
plt.show()
```
