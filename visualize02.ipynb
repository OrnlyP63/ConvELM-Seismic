```python
from obspy.clients.syngine import Client
client = Client()
```

```python
st = client.get_waveforms(model="ak135f_5s", network="IU", station="ANMO",
                          eventid="GCMT:C201002270634A")
```

```python
st.plot()
```

```python
st[0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(st[0].data)
```

```python
tr = st[0]
```

```python
tr.times("matplotlib")
```

```python
fig = plt.figure(figsize=(10, 5), dpi=80)
ax = fig.add_subplot(111)
s = st[0]
sp = s[2800:5000]
ss = s[5000:8100]
sb = s[8100:13400]
time_axis = s.times("matplotlib")
ax.plot(time_axis, s,  c='k')
ax.plot(time_axis[2800: 5000],sp, c='r', label='P-wave')
ax.plot(time_axis[5000: 8100],ss, c='g', label='S-wave')
ax.plot(time_axis[8100: 13400],sb, c='b', label='Surface-wave')
ax.grid()
ax.xaxis_date()
ax.set_xlabel('Time', fontdict={'size': 24})
ax.set_ylabel('Ground acceleration (Gal)', fontdict={'size': 24})
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=16)
#ax.legend(prop={'size': 18})
plt.savefig('wave.png')
plt.show()
```
